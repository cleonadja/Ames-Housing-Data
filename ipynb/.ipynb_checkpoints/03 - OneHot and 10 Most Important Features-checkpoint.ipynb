```python
def apply_scale(dataframe, scaling_function):
    numerical_df = dataframe.select_dtypes(include=[float])
    print(numerical_df.columns)
    numerical_df = scaling_function(numerical_df)
    tmp_df = dataframe.copy()
    tmp_df[numerical_df.columns] = numerical_df
    return tmp_df

def gelman_scale(dataframe):
    return (dataframe - dataframe.mean())/(2*dataframe.std())

def standard_scale(dataframe):
    return (dataframe - dataframe.mean())/(dataframe.std())

df_gelman = apply_scale(housing_df, gelman_scale)
df_standard = apply_scale(housing_df, standard_scale)
```

```python
fig = plt.figure(figsize=(20,9))
for i, col in enumerate(housing_df.iloc[:,0:10]):
    fig.add_subplot(3,11,1+i)
    sns.distplot(housing_df[col])
    plt.xlim(-3,3)
    fig.add_subplot(3,11,12+i)
    sns.distplot(df_gelman[col])
    plt.xlim(-3,3)
    fig.add_subplot(3,11,23+i)
    sns.distplot(df_standard[col])
    plt.xlim(-3,3)
```

## 10 Most Important Features

```python
housing_df.corr()
```

```python
housing_corr_df = housing_df.corr()

housing_SalePrice_corr = housing_corr_df.SalePrice

housing_SalePrice_corr_abs = housing_SalePrice_corr[np.abs(housing_SalePrice_corr) > 0.5]
housing_SalePrice_corr_abs
```

```python
housing_SalePrice_corr_abs.index
```

```python
housing_df_reshape = housing_SalePrice_corr_abs.values.reshape(13, 1)
housing_df_reshape
```

```python
corr = housing_df_reshape
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask, 0)] = True
with sns.axes_style("white"):
    ax = sns.heatmap(corr, mask=mask, square=True, annot=True,
                     cmap='RdBu', fmt='+.3f')
    plt.xticks(rotation=45, ha='center')
```

Based on the above correlation the 10 most important features to predict the SalePrice are 'YearBuilt', 'TotalBsmtSF', 'FirstFlrSF', 'GrLivArea', 'GarageArea', 'ExterQual_TA', 'BsmtQual_Ex', 'FullBath_1', 'KitchenQual_TA', 'GarageCars_3'.
