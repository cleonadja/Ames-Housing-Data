## Distribution Plot Showing the Distribution, Mean, and Median

```r
multiplot(hist_with_kde(LotFrontage),
          hist_with_kde(LotArea), 
          hist_with_kde(FirstFlrSF),
          hist_with_kde(SecondFlrSF),
          cols = 4)
multiplot(hist_with_kde(PoolArea),
          hist_with_kde(YrSold), 
          hist_with_kde(GarageArea),
          hist_with_kde(TotalBsmtSF),
          cols = 4)
```

These KDE plots show the distribution of a few features. The LotArea for example has most values between 0 and 250k squarefeet with a few houses having a larger lot size than that. The linear feet of street connected to the property varies mostly in the range of 0-125 feet. First and second floor squarefootage are distributed very differently. While the first floor squarefootage is somewhere between 0-2500 we are seeing the second floor squarefootage spanning between 300-1250 squarefoot - however, most houses do not have a second floor or a pool. The year of the house sold is fairly evently distributed across the 5 years reprsenetd in this dataset (2006-2010). The average garage area is about 500 squarefeet almost no garages under 200 sq and a few larger than 100sq. The average basement squarefootage is about 100o sq.

## Distribution Plot Showing the Distribution of a Numerical Feature Broken Down By a Categorical Feature

```r
hist_with_kde_numerical_by_category(SalePrice, as.factor(YrSold))
```

The avbove plot is showing the sale price of a home by the year sold and we can see that the sale price is fairly consistent across the years represented in the dataset (2006-2010). The year a house is being sold does not have an impact on the sale price.

```r
multiplot(hist_with_kde_numerical_by_category(SalePrice,HouseStyle),
          hist_with_kde_numerical_by_category(SalePrice,ExterQual), 
          hist_with_kde_numerical_by_category(SalePrice,Street),
          cols = 3)
```

Above we can see that the house style does impact the sale price to a small degree while the exterior quality is highly correlated with the sale price as well as the type of road access.
